```python
import csv
import os
import pandas as pd
```

```python
filename = os.path.join(os.getcwd(), 'Prepared_Data.csv')
df = pd.read_csv(filename, header = 0)

df.head()
```

```python
# Renaming "Respiratory" column:
df = df.rename(columns={'Resipiratory':'Respiratory'})
df['Respiratory']
```

```python
# Dropping "City", "State" columns:
df = df.drop(columns=['City','State'], axis = 1)
df.head()
```

```python
# Creating new df for each category: 
df_resp = df.copy()
df_repr = df.copy()
```

```python
# Removing all other categories from df_resp:
drop_resp = ['Diagnoses','Eyes, Ears, Nose, & Throat','Urinary', 'Digestive','Circulatory', 'Skin', 'Brain', 'Nervous', 'Immune', 'Reproductive'] 
df_resp = df_resp.drop(columns=drop_resp)
df_resp.head()
```

```python
# Removing all other categories from df_repr:
drop_repr = ['Diagnoses','Eyes, Ears, Nose, & Throat','Urinary', 'Digestive','Circulatory', 'Skin', 'Brain', 'Nervous', 'Immune', 'Respiratory'] 
df_repr = df_repr.drop(columns=drop_repr)
df_repr.head()
```

```python
# Creating new CSV files:
df_resp.to_csv('Respiratory_Symptoms.csv', index = False)
df_repr.to_csv('Reproductive_Symptoms.csv', index = False)
```
