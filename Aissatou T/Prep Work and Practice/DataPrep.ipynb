```python
def random_centroids(data, k):
    centroids=[]
    for i in range (k):
        centroid = centroid = data.apply(lambda x: float (x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

```python
def get_labels (data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data- x) ** 2).sum(axis = 1)))
    return distances.idxmin(axis=1)
    
```

```python
labels = get_labels(data, centroids)
labels.value_counts()
```

```python
def new_centroids (data, labels, k):
    return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters (data, labels, centroids, iteration):
    pca =PCA(n_components = 2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration{iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels, cmap='viridis')  
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1], c='red', marker='X')  

    plt.show()
```

```python
max_iterations=100
k=3

centroids = random_centroids(data, k)
old_centroids= pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids= centroids
    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, k)
    plot_clusters(data, labels, centroids, iteration)
    iteration += 1
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Normalize your data 
data = (data - data.mean()) / data.std()

# Use sklearn's KMeans
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=max_iterations, n_init=10, random_state=0)
labels = kmeans.fit_predict(data)

# PCA for visualization
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data)

# Print explained variance
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance (2 components): {explained_variance.sum()}")

# Plot the clusters
plt.figure(figsize=(10, 6))
plt.title(f'K-means Clustering with PCA (2 components)')
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label="Cluster Centers")
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is your normalized dataset
pca = PCA().fit(data)

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plotting the Cumulative Summation of the Explained Variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_explained_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')  # for each component
plt.title('Explained Variance vs. Number of Components')
plt.axhline(y=0.80, color='r', linestyle='--', label="80% Explained Variance")
plt.axhline(y=0.90, color='g', linestyle='--', label="90% Explained Variance")
plt.legend()
plt.show()

# Finding the number of components for desired explained variance
n_80 = np.where(cumulative_explained_variance > 0.80)[0][0] + 1  # +1 as indexing starts from 0
n_90 = np.where(cumulative_explained_variance > 0.90)[0][0] + 1

print(f"Number of components to capture 80% variance: {n_80}")
print(f"Number of components to capture 90% variance: {n_90}")
```
